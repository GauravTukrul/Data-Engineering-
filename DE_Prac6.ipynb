Practical 6: Perform Regression using ML Pipeline with PySpark in Databricks on US cars dataset


```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml import Pipeline
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
data = [
    (2015, 50000, "Toyota", "Petrol", 15000),
    (2018, 30000, "Honda", "Diesel", 18000),
    (2017, 40000, "Ford", "Petrol", 16000),
    (2020, 20000, "Toyota", "Diesel", 22000),
    (2016, 45000, "Honda", "Petrol", 14000)
]
columns = ["year", "mileage", "brand", "fuel", "price"]
df = spark.createDataFrame(data, columns)

df.show()
```

```python
brand_indexer = StringIndexer(inputCol="brand", outputCol="brand_index")
fuel_indexer = StringIndexer(inputCol="fuel", outputCol="fuel_index")
```

```python
assembler = VectorAssembler(
    inputCols=["year", "mileage", "brand_index", "fuel_index"],
    outputCol="features"
)

```

```python
lr = LinearRegression(featuresCol="features", labelCol="price")
```

```python
pipeline = Pipeline(stages=[brand_indexer, fuel_indexer, assembler, lr])
```

```python
model = pipeline.fit(df)
```

```python
result = model.transform(df)
result.select("year", "mileage", "brand", "fuel", "price", "prediction").show()
```
