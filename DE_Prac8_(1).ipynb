# Practical 8 :	Perform Deep Learning in Keras in Distributed computing using PySpark


```python
#Step 1: Install Required Libraries
! pip install pyspark tensorflow tensorflowonspark

#Step 2 :
from pyspark.sql import SparkSession
from tensorflowonspark import TFCluster
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
# 1. Model Function
# -----------------------------
def model_fun(args, ctx):
    import tensorflow as tf
    from tensorflow import keras
    from tensorflow.keras import layers
    import numpy as np


    # Dummy dataset
    x = np.random.rand(1000, 2)
    y = (x[:, 0] + x[:, 1] > 1).astype(int)


    # Model
    model = keras.Sequential([
        layers.Dense(8, activation='relu', input_shape=(2,)),
        layers.Dense(4, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])


    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])


    model.fit(x, y, epochs=5, batch_size=32)


    print("Training done on worker")
```

```python
# 2. Spark Session
# -----------------------------
spark = SparkSession.builder \
    .appName("Distributed Keras") \
    .getOrCreate()


sc = spark.sparkContext

```

```python

# 3. Start Cluster

cluster = TFCluster.run(
    sc,
    model_fun,
    {},
    num_executors=2,
    num_ps=1,
    input_mode=TFCluster.InputMode.SPARK
)
```

```python
# 4. Shutdown Cluster
# -----------------------------
cluster.shutdown()


print("Distributed Training Completed")


spark.stop()
```
