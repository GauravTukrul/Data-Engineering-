Practical 4 : Data Cleaning and Transformation Using ML Pipeline.

```python
import pandas as pd
import numpy as np

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
data = {
    'Age': [25, np.nan, 35, 40, np.nan],
    'Salary': [50000, 60000, np.nan, 80000, 75000],
    'Department': ['HR', 'IT', 'Finance', np.nan, 'IT']
}
```

```python
df = pd.DataFrame(data)
print("Original Data:\n", df)
```

```python
numeric_features = ['Age', 'Salary']
categorical_features = ['Department']
```

```python
# Numeric pipeline → fill missing + scale
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Categorical pipeline → fill missing + encode
categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder())
])
```

```python
preprocessor = ColumnTransformer([
    ('num', numeric_pipeline, numeric_features),
    ('cat', categorical_pipeline, categorical_features)
])
```

```python
cleaned_data = preprocessor.fit_transform(df)
print("\nCleaned and Transformed Data:\n", cleaned_data)
```
