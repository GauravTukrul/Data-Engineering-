```python
from pymongo import MongoClient
```

```python
client = MongoClient("mongodb://localhost:27017/")
print("Connected to MongoDB")
```

```python
db = client["college_db"]
collection = db["students"]
print("Database and Collection ready")
```

```python
student = {
    "name": "Piyali",
    "course": "Data Science",
    "age": 22,
    "marks": 85
}

result = collection.insert_one(student)
print("Inserted ID:", result.inserted_id)
```

```python
students = [
    {"name": "Amit", "course": "AI", "marks": 88},
    {"name": "Riya", "course": "ML", "marks": 92},
    {"name": "Kunal", "course": "Cloud", "marks": 80}
]

result = collection.insert_many(students)
print("Inserted IDs:", result.inserted_ids)
```

```python
print("\nAll Documents:")
for doc in collection.find():
    print(doc)
```

```python
query = {"name": "Piyali"}
result = collection.find_one(query)
print("\nFound Document:", result)
```

```python
query = {"name": "Piyali"}
new_values = {"$set": {"marks": 90, "age": 23}}

result = collection.update_one(query, new_values)
print("Documents updated:", result.modified_count)
```

```python
query = {"name": "Kunal"}
result = collection.delete_one(query)
print("Documents deleted:", result.deleted_count)
```

```python
print("\nAfter Deletion:")
for doc in collection.find():
    print(doc)
```

```python
result = collection.delete_many({})
print("All documents deleted:", result.deleted_count)
```

```python
collection.drop()
print("Collection dropped")
```

```python
client.drop_database("college_db")
print("Database dropped")
```
